# Simple Line Plots

```python
# Set up the notebook
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# Start by creating a figure and axes
fig = plt.figure()
ax = plt.axes()

# Now plot some data
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x));

# ";" supresses this message from showing -> [<matplotlib.lines.Line2D at 0x7fd51047c490>]
```

```python
# Create a single figure with multiple lines using PyLab interface (two interfaces discussed in chapter 25)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

### Adusting the Plot: Line Colors and Styles

```python
# Line colors
plt.plot(x, np.sin(x - 0), color='blue')         # by name
plt.plot(x, np.sin(x - 1), color='g')            # color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')         # grayscale 0 - 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')      # hex code (RRGGBB, 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3))  # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse');  # HTML color names 
```

```python
# Line styles
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');


# For short, you can use the following codes:
# plt.plot(x, x + 0, linestyle='-')   # solid
# plt.plot(x, x + 1, linestyle='--')  # dashed
# plt.plot(x, x + 2, linestyle='-.')  # dashdot
# plt.plot(x, x + 3, linestyle=':')   # dotted
```

### Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
# Reverse y axis (if you wanted to do this for whatever reason)
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

```python
# Tight layout
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
# Equal layout
plt.plot(x, np.sin(x))
plt.axis('equal');
```

### Labeling plots

```python
# see page 242 ax.set (another, and potentially better, way to do this)
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-', 'g', label='sin(x)')
plt.plot(x, np.cos(x), ':', 'b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

### Matplotlib Gotchas
Most plt functions translate directly to ax mothods (plt.plot -> ax.plot, plt.legent -> ax.legent, etc.), this is not the case for all commands. 

In particular, functions to set limits, labels, and titles are slightly modified

plt.xlabel -> ax.set_xlabel <br>
plt.ylabel -> ax.set_ylabel <br>
plt.xlim -> ax.set_xlim <br>
plt.ylin -> ax.set_ylim <br>
plt.title -> ax.set_title <br>

```python
# In the object-oriented interface to plotting, rather than calling these functions individually,
# it is often more convenient to use the ax.set method to set all these properties at once

ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
      xlabel='x', ylabel='sin(x)',
      title='A Simple Plot');
```
